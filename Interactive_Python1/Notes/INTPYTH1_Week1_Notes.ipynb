#### *Video 1.3 Arithmetic Expressions*

**A. Intro**

- the ```int()``` function will not round...it will just chop off everything after the decimal point
- floating point numbers are only an *approximation* of real, decimal-notation numbers
- if one of the operands is a float, the answer is a float
- if all operands are integers, the answer is an int
- the division operator works differently in Python 2 than in Python 3
    - if both operands are integers, the answer to a division problem will be rounded up

````python
print 1/3, 5/2, -7/3, -9/4
````

- Order of operations:
    - Parentheses, exponents, multiplication, division, addition, subtraction
- when in doubt, always just group stuff in parentheses

#### *Video 1.4 Variables*

**A. Intro to Variables**

- variables = containers that store the result of calculations
- think of variables as tools to make your program more efficient and more understandable
- typically use underscore variables
- variable names have to start with letters or underscores
- multiname names must be connected with underscores
- Single equal, ```=```, is used for assignment.
    - Dobule equal, ```==```, is used for a logical equality test
- To increment an existing variable, two equivalent options:
    - ```my_age = my_age + 1```
    - ```my_age += 1```
- Give your variables memorable names...it will help later
