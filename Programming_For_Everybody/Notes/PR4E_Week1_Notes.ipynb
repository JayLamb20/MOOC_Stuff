####*Video 1.1 Why Program*

**A. Computer want to be helpful...**

- computers do stuff for us
- we need to speak their language to get them to do what we want them to do
- each button/bit of software is saying "what do you want me to do next?"
- programmers are the people who speak the language of the computer

**B. Being a Programmer**

- move from the outside (users clicking buttons) to inside (programs controlling computer tools)
- software connects data, hardware, networks

**C. Why be a programmer?**

- to get some task done - we are the user and the programmer
    - clean up survey data
- produce something for others to use - a programming job
    - fix a performance problem in the Sakai software
    - add guestbook to a webpage
    
**D. What is code?**

-  a sequence of store instructions
- "a little piece of our intelligence in the computer"
- a piece of creative art - particularly when we do a good job on user experience
- people can use fuzzy logic, correct little mistakes (computers need explicity, perfect instructions)
- computers are better at "concentration" than humans
    - able to scale really fast
- you don't have to do menial work if you can learn to tell the computer to do it
